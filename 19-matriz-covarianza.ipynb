# <span style="color:royalblue"><center> Matriz de Correlación</center></span>

Medir las desviaciones o variaciones d euna variable x con relación a otra variable. Esto además está relacionado con el cambio que tiene la curva en el sentido estricto de una primera derivada. 

```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.get_dataset_names()
```

```python
sns.pairplot(iris)
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
)
scaled.T
```

```python
covariance_matrix = np.cov(scaled.T)
covariance_matrix
```

```python
plt.figure(figsize=(10,10))
sns.set(font_scale=1.5)
hm = sns.heatmap(covariance_matrix,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 annot_kws={'size': 12},
                 yticklabels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                 xticklabels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
```
