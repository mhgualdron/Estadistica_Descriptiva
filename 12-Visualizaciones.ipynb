Hay muchos tipos de gráficas que son posibles verlos en páginas donde resumen de muy buena manera los atributos, aplicaciones y utilidades que tiene cada uno de los distintos tipos de gráficos; una muy buena página para poder ver ésto es la página de [DataViz](https://datavizproject.com/).

[Seaborn documentation](https://seaborn.pydata.org/generated/seaborn.scatterplot.html)

```python
import pandas as pd 
import seaborn as sns

iris = sns.load_dataset('iris')

iris.head()
```

```python
# scatterplot 
sns.scatterplot(data=iris, x = 'sepal_length', y = 'petal_length', hue = 'species')
```

http://seaborn.pydata.org/generated/seaborn.jointplot.html

```python
# joint plot
sns.jointplot(data=iris, x = 'sepal_length', y = 'petal_length',hue='species')
```

https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
# box plot 
sns.boxplot(x = 'species', y = 'sepal_length', data = iris)
```

https://seaborn.pydata.org/generated/seaborn.barplot.html

```python
#Barplot
sns.barplot(x = 'species', y = 'sepal_length', data = iris)
```
